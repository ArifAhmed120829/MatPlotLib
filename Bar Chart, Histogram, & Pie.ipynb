```python
Company=['Google','Amazon','Facebook','Twitter']
Revenue=[173,386,85.96,30.4]
Profit = [48,2,34,12]
```

```python
ypos = np.arange(len(Company))
ypos
```

```python
#numpy module er attribute aida arange
#String take numbare porinoto korse
```

```python
plt.bar(ypos -0.2,Revenue,width = 0.4,label = "Revenue")
plt.bar(ypos +0.2,Profit,width = 0.4, label = "Profit")
plt.xticks(ypos,Company)#xticks function
plt.ylabel("Revenue(billion dollar)")
plt.xlabel("Company name")
plt.title("Most Income")
plt.legend()
plt.show()
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_woman = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_woman], bins = [80,100,125,150], rwidth= 0.95, color=['green','orange'], label = ['men','women'])
plt.legend()
plt.show()
```

```python
#blood sugar 80 to 100 normal , 100 t0 125 prediabetic & 125+ diabetic
```
